## Analysis On the basis of Age

```python
ax= sns.countplot(x="Age Group",data = df,hue="Gender")
plt.title('Age Vs Count')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
x=df.groupby(df["Age Group"],as_index=False).sum()[["Age Group","Amount"]]
x
```

```python
ax=sns.barplot(x="Age Group",y="Amount", data=x)
plt.title("Age Vs Amount")
plt.figure(figsize=(1,100))
```

#### Ans : Most of the buyers from age group 26 TO 35

## Analysis On the basis of State

```python
df.head()
```

```python
State_Sales=df.groupby(df["State"],as_index=False)["Orders"].sum().sort_values(by = "Orders",ascending=False).head(10)
State_Sales
```

```python
plt.figure(figsize=(17,7))
sns.barplot(data = State_Sales, x = 'State',y= 'Orders')
plt.title("Top 10 Sate by Sales")
```

### Ans: Most of the sales are from Up,Maharashtra and Karnatka

## Analysis on the basis of Martial Status

```python
marital_data = df.groupby(df["Marital_Status"],as_index=False).count()[["Marital_Status","User_ID"]]
marital_data
```

```python
plt.bar(["Unmarried","Married"],marital_data["User_ID"],color=["Blue","Orange"])
plt.xlabel('Marital Staus')
plt.ylabel('Count')
plt.title('Marital Status and Their Count')
plt.show()
```

```python
marital_data=df.groupby(["Marital_Status","Gender"],as_index=False).sum()
marital_data
```

```python
ax=sns.barplot(data= marital_data,x="Marital_Status",y="Amount",hue= "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

### Ans : Most of the buyer are married women and their purchasing power are also high

## Analysis on the basis of Occupation

```python
occupation_count=df.groupby("Occupation",as_index=False).count()[["Occupation","User_ID"]]
occupation_count
```

```python
plt.figure(figsize=(20,5))
ax=sns.barplot(data=occupation_count,x="Occupation",y="User_ID")
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation and their Count')
plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

#### Ans : Most of the buyers are from IT sector,Healthcare and Aviation background

## Analysis on the basis of product category

```python
plt.figure(figsize=(25,5))
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(20,5))
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

### Ans : Most of the sales product are food,clothing and electronics gadgets

## Final Conclusion

### Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
